### Exercise 1: 

Write a function called writeNLetters() that creates a file named “words.txt” and writes the English alphabet by a specified number of letters on each line. For example, when writeNLetters() is called with an argument value of 3 (i.e., writeNLetters(3)), it should create the following content in “words.txt” file.


ABC
DEF
GHI
JKL
MNO
PQR
STU
VWX
YZ

```python
# Your answer for Exercise 1
def writeNLetters(x):
    alphabet="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    last=26%x
    if last!=0:
        #not divisior of 26
        lastrings=alphabet[-last:]
        firststrings=alphabet[:-last]
        
        with open("words.txt","+w") as file:
            for i in range(int(26/x)):
                file.write(firststrings[:x])
                file.write(" ")
                firststrings=firststrings[x:]
    
            file.write(lastrings)
            
    else:
        #26s divisior
        with open("words.txt","+w") as file:
            for i in range(int(26/x)):
                file.write(alphabet[:x])
                file.write(" ")
                alphabet=alphabet[x:]
            
writeNLetters(3)
```

```python
# Your answer for Exercise 2
#

def get_grade(grade_str):
    grade_dict = {
        "A": 4.00,
        "A-": 3.70,
        "B+": 3.30,
        "B": 3.00,
        "B-": 2.70,
        "C+": 2.30,
        "C": 2.00,
        "C-": 1.70,
        "D+": 1.30,
        "D": 1.00,
        "F": 0.00,
    }

    return grade_dict[grade_str]


with open("student_grades.csv") as file:
    firstlistoflines=file.readlines()
lastloflines=[]
for line in firstlistoflines[1:]:
    newline=line[:-1]
    lastline=newline.split(";")
    lastloflines.append(lastline)

for line in range(len(lastloflines)):
    for grade in range(1,6):
        
        lastloflines[line][grade]=get_grade(lastloflines[line][grade])
#testprint(lastloflines)

dictwithaverage={}
for line in lastloflines:
    
    average=(line[1]*6+line[2]*4+line[3]*6+line[4]*8+line[5]*4)/28
    averager=round(average,2)
    dictwithaverage[line[0]]=averager
    
#testprint(dictwithaverage)
with open("student_gpas.csv","+w") as file:
    file.write("Student ID;GPA\n")
    for number in dictwithaverage:
        file.write(number+" "+str(dictwithaverage[number])+"\n")
    
    

    
```

```python
# Your answer for Exercise 3

import os

def add_contact(file_name):
    with open(file_name, 'a') as file:
        name = input("Enter contact's name: ")
        email = input("Enter contact's email address: ")
        phone = input("Enter contact's phone number: ")
        file.write(f"{name}, {email}, {phone}\n")
        print("Contact added successfully!")

def update_contact_email(file_name):
    name_to_update = input("Enter contact's name whose email you want to update: ")
    new_email = input("Enter new email address: ")

    with open(file_name, 'r') as file:
        lines = file.readlines()

    with open(file_name, 'w') as file:
        for line in lines:
            contact_info = line.strip().split(", ")
            if contact_info[0] == name_to_update:
                contact_info[1] = new_email
                line = f"{contact_info[0]}, {contact_info[1]}, {contact_info[2]}\n"
            file.write(line)

    print("Email address updated successfully!")

def display_contact_details(file_name):
    name_to_display = input("Enter contact's name to display details: ")

    with open(file_name, 'r') as file:
        for line in file:
            contact_info = line.strip().split(", ")
            if contact_info[0] == name_to_display:
                print(f"Name: {contact_info[0]}")
                print(f"Email: {contact_info[1]}")
                print(f"Phone: {contact_info[2]}")
                break
        else:
            print("Contact not found.")

def main():
    file_name = "address_book.txt"

    while True:
        print("\nAddress Book Management System")
        print("1. Add a new contact")
        print("2. Update a contact's email")
        print("3. Display a contact's details")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            add_contact(file_name)
        elif choice == '2':
            update_contact_email(file_name)
        elif choice == '3':
            display_contact_details(file_name)
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
